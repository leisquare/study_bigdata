```python
import numpy as np #csv파일을 numpy배열로 읽기
import tensorflow.keras.utils as utils # 라벨링
from tensorflow.keras.models import Sequential # 모델생성
from tensorflow.keras.layers import Dense, Activation #입력값, 출력값, activation function지정
```

# 1. 데이터 셋 생성

```python
dataset=np.loadtxt('data/pima-indians-diabetes.csv',delimiter=',')
dataset
```

```python
#학습용 데이터(훈련데이터)
x_train=dataset[:700,:8]
y_train=dataset[:700,8]

#테스트 데이터
x_test=dataset[700:,:8]
y_test=dataset[700:,8]
```

```python
x_train.shape,y_train.shape
```

```python
x_test.shape,y_test.shape
```

```python
# 라벨링 0->[1,0]/1 ->[0,1]
```

# 2. 모델 구성하기

```python
model = Sequential()
# input_dim=8 최중 출력 1
model.add(Dense(128,input_dim=8, activation="relu")) 
model.add(Dense(64,activation="relu"))
model.add(Dense(16,activation="relu"))
model.add(Dense(1,activation="sigmoid")) #원핫인코딩안했으면이거. 했으면 softmax
```

# 학습과정 설정

```python
model.compile(loss = "binary_crossentropy",optimizer="adam",metrics=['accuracy'])
```

# 학습시키기

```python
hist = model.fit(x_train,y_train,epochs=1500,batch_size=64)
```

# 5. 모델 평가하기

```python
score=model.evaluate(x_test,y_test)
```

```python
print("loss: %s \naccuracy: %.2f" % (score[0],score[1]))
```
