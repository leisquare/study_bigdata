# XOR

```python
import numpy as np
from tensorflow.keras.models import Sequential #모델 객체 생성시
from tensorflow.keras.layers import Dense #레이어 쌓을 때 
```

```python
# 1. data set
x_data=[[0,0],
       [0,1],
       [1,0],
       [1,1]]
y_data=[[0],[1],[1],[0]]

# 2. 모델 구성
model=Sequential()
model.add(Dense(10,input_dim=2,activation="relu"))
model.add(Dense(20,activation="relu")) #은닉층은 input_dim 쓰지않음
model.add(Dense(10,activation="relu")) 
model.add(Dense(1,activation="sigmoid")) 

# 3. 모델 학습과정
model.compile(loss="mse",optimizer = "adam",metrics=['binary_accuracy']) #알아서 로스 줄임
#model.compile(loss="crossentropy",optimizer = "adam",metrics=['binary_accuracy'])
#model.compile(loss="binary_crossentropy",optimizer = "adam",metrics=['binary_accuracy'])

# 4. 학습하기
print(model.summary()) #파라미터의 갯수
fit_hist=model.fit(x_data,y_data,epochs=100,verbose=1)
```

```python
fit_hist.history.keys()
```

```python
for acc in fit_hist.history['binary_accuracy']:
    print(acc,end='\t')
```

```python
for acc in fit_hist.history['loss']:
    print(acc,end='\t')
```

```python
import matplotlib.pyplot as plt
plt.plot(fit_hist.history['loss'])
plt.show()
```

```python
# 입력 스타일은 0 0 
while True:
    input_list=input("space로 분리해서 0이나 1을 2개 입력(종료:9)").strip().split()
    input_data=list(map(int,input_list))
    print(input_list)
    if input_data[0] == 9:
        break;
    input_data=np.array(input_data).reshape(1,2)
    print('입력값:',input_data)
    print('출력값:',int(model.predict(input_data).round()))
```

# OR

```python
# 1. data set
x_or_data = np.array([[0,0],[0,1],[1,0],[1,1]], 'float32')
y_or_data = np.array([[0],[1],[1],[1]] , 'float32')
# 2. 모델 구성
model = Sequential()
model.add(Dense(32, input_dim=2, activation='relu'))
model.add(Dense(1, activation="sigmoid"))
# 3. 모델 학습과정
model.compile(loss="mse", optimizer="adam", metrics=['binary_accuracy'])
# model.compile(loss="categorical_crossentropy", optimizer="adam", metrics=['accuracy'])
# model.compile(loss="binary_crossentropy", optimizer="adam", metrics=['accuracy'])
# 4. 학습
fit_hist = model.fit(x_or_data, y_or_data, epochs=300, verbose=1)
```

```python
plt.plot(fit_hist.history['loss'])
plt.show()
```

# AND

```python
# 1. data set
x_and_data=[[0,0],
       [0,1],
       [1,0],
       [1,1]]
y_and_data=[[0],[0],[0],[1]]

# 2. 모델 구성
model=Sequential()
model.add(Dense(32,input_dim=2,activation="relu"))
model.add(Dense(1,activation="sigmoid")) 

# 3. 모델 학습과정
model.compile(loss="mse",optimizer = "adam",metrics=['binary_accuracy']) #알아서 로스 줄임
#model.compile(loss="crossentropy",optimizer = "adam",metrics=['binary_accuracy'])
#model.compile(loss="binary_crossentropy",optimizer = "adam",metrics=['binary_accuracy'])

# 4. 학습하기
print(model.summary()) #파라미터의 갯수
fit_hist=model.fit(x_and_data,y_and_data,epochs=500,verbose=1)
```

```python
plt.plot(fit_hist.history['loss'])
```

```python
#1000회로 늘리기
fit_hist = model.fit(x_and_data, y_and_data, epochs=1000)
plt.plot(fit_hist.history['loss'])
```
