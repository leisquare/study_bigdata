```python
#종속변수분할
Input=raw_data.iloc[:,:-1]
Target=raw_data.iloc[:,[-1]] 
Target
```

```python
#스케일 조절
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(Input)
scaled_input=scaler.transform(Input)
scaled_input=pd.DataFrame(scaled_input)
scaled_input
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(scaled_input,Target,test_size=0.3,random_state=5)
print(X_train.shape,X_test.shape,Y_train.shape,Y_test.shape)
```

```python
#원핫
Y_train = utils.to_categorical(Y_train)
Y_test  = utils.to_categorical(Y_test)
Y_test
```

```python
# 모델 구성
model = Sequential()
model.add(Dense(units=50,input_dim=11,activation='relu'))
#과적합을 해결하고자 기억하되 두루뭉실하게 기억하기
model.add(Dense(units=30,activation='relu'))
model.add(Dense(units=20,activation='relu'))
model.add(Dense(units=9,activation='softmax'))
#여러개로나오니까 softmax, 그리고 9종으로 분류되니 units=9 결과물이 9개 나온다는걸....흑
model.summary()
```

```python
#학습과정설정
model.compile(loss='mse',optimizer='adam',metrics=['accuracy'])
#학습시키기
fit_hist = model.fit(X_train,Y_train,epochs=100,batch_size=50,verbose=2) 
```
