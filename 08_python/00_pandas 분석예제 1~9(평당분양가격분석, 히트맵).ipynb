# 4. df_last 데이터셋에 평당분양가격 컬럼을 추가

```python
df_last["분양가격(㎡)"] = pd.to_numeric(df_last["분양가격(㎡)"], errors="coerce")
df_last["평당분양가격"] = df_last["분양가격(㎡)"] * 3.3
df_last
```

# 5. df_last 데이터셋에 전용면적 컬럼 추가
(규모구분 컬럼을 이용할 것)

```python
#df_last.loc[:,["규모구분","전용면적"]]
```

```python
df_last1=df_last

df_last1["전용면적"]= df_last1["규모구분"].apply(lambda x:(x.replace("이하","")))
df_last1["전용면적"]= df_last1["전용면적"].apply(lambda x:(x.replace("전용면적","")))
df_last1["전용면적"]= df_last1["전용면적"].apply(lambda x:(x.replace("초과","")))
df_last1["전용면적"]= df_last1["전용면적"].apply(lambda x:(x.replace("㎡ ","㎡~")))
df_last1
```

# 6.	사용하지 않을 컬럼은 제거
#(df_last.drop([‘규모구분’,’분양가격((㎡)’], axis=1 이용),
#메모리 사용량을 전후로 확인한다.



```python
df_last1=df_last1.drop(['규모구분','분양가격(㎡)'], axis=1)
df_last1
```
