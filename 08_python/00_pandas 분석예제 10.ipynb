# 1-1. import 하고, melt, concat, pivot, transpose와 같은 reshape가능을 활용해 분석


```python
#1-1-1) import
import pandas as pd
df_first=pd.read_csv('E:/kim_jiwon/src/08_python/data/전국 평균 평당 분양가격(2013년 9월부터 2015년 8월까지).csv',sep=',',encoding='cp949')
df_first.head(10)
```

```python
df_last=pd.read_csv('E:/kim_jiwon/src/08_python/data/주택도시보증공사_전국 신규 민간아파트 분양가격 동향_20210215.csv',sep=',',encoding='cp949')
df_last.head(10)
```

```python
#df_first 데이터 정제
df_first = df_first.melt(id_vars="지역", 
                          var_name="연도월", 
                          value_name="평당분양가격")
df_first.head()
```

```python
df_first["연도"]= df_first["연도월"].apply(lambda x:int(x.split("년")[0]))
df_first["월"]= df_first["연도월"].apply(lambda x:int(x.split("년")[1].replace("월","")))
```

```python
df_first=df_first.drop(['연도월'],axis=1)
```

```python
df_first.head()
```

```python
#df_last 데이터 정제
df_last["분양가격(㎡)"] = pd.to_numeric(df_last["분양가격(㎡)"], errors="coerce")
df_last["평당분양가격"] = df_last["분양가격(㎡)"] * 3.3

df_last
```

```python
#컬럼네임변경
df_last['지역']=df_last['지역명']
df_last
```

```python
# 모든 면적만 남기고 규모구분 제거, 컬럼 이름 통일
df_last=df_last=df_last.loc[df_last['규모구분']=="모든면적",["지역","연도","월","평당분양가격"]]
df_last
```

```python
df_first=df_first.loc[:,['지역','연도','월','평당분양가격']] 
df_first
```

```python
df_ex=pd.concat([df_first,df_last])
df_ex.head(50)
```

```python
#분석 개시 전, 결측치 확인
df_ex.info()
```
