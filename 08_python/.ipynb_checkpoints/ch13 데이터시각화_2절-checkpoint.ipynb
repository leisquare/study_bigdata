```python
#시작 전 설정
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline 
%config InlineBackend.figure_format='retina'
#원하는 한글폰트 지정
plt.rc("font",family="Malgun Gothic")
plt.rc("axes",unicode_minus=False) #Y축
#경고메세지 안보이게
import warnings
warnings.filterwarnings(action='ignore')
```

4) 축과 눈금

```python
import numpy as np
x = np.linspace(0,10,100)
y=np.cos(x)
fig, axes = plt.subplots(2,2,figsize=(12,5))
#plt.subplots_adjust로 서브플롯간의 간격을 변경한다.
plt.subplots_adjust(hspace=0.4,wspace=0.8) #hspace는 위아래 wspace는 옆여백
axes[0,0].scatter(x,y,marker='.')
axes[0,0].set(title="An Example Axes",ylabel="y-axis",xlabel="엑스축",
             xlim=[0,12],ylim=[-2,2]) #축 변경
axes[0,1].scatter(x,y,marker='*',c='b')
axes[0,1].set_xlim(2,6)
axes[0,1].set_xlabel('x축 0~5')
axes[0,1].set_ylim(-1.5,1.5)
axes[0,1].set_ylabel('y축 -1.5~1.5')
axes[1,0].scatter(x,y,marker='^',c='c')
#x축 눈금 위치
axes[1,0].set_xticks(range(0,11,3))
axes[1,0].set_xticklabels([-100,3,'6개','foo'])
axes[1,0].set_yticks([-2,0,1,2])
axes[1,0].set_yticklabels(['월','화',1,'Max'])
axes[1,1].scatter(x,y,c='r')
axes[1,1].set(xticks=[0,2.5,5,7.5,10], xticklabels=['min',2.5,5,7.5,'Max'],title='test')
axes[1,1].grid(True)
axes[1,1].spines["top"].set_visible(False)
axes[1,1].spines["bottom"].set_position(("outward",10))
plt.show
```

5) 축 공유

```python
import numpy as np
x = np.arange(0,10)
y1=0.5*x**2 #0.5*(x**2)
y2=-1*y1
```

```python
fig, ax1=plt.subplots()
ax1.plot(x,y1,'g^:')
ax1.set(xlabel='X data',ylabel='Y1 data')
ax2 = ax1.twinx() #ax1과 x축을 공유하는 ax2
ax2.plot(x,y2,'b--')
ax2.set_ylabel('y2 data',color='b')

ax3 = ax1.twiny() #ax1과 x축을 공유하는 ax3
ax3.plot(-x,y1,'ro-')
ax3.set_xlabel('x data',color='r')

plt.show()
```

7) 범례표시

```python
x=np.arange(0,10)
y=0.2*x**2
```

```python
fig,ax=plt.subplots(figsize=(8,4))
ax.plot(x,y,'g',label='GREEN')
ax.plot(x,-y,'b',label="BLUE")
ax.legend()
plt.legend(loc='center right',bbox_to_anchor=(1.25,0.5),ncol=1) #범례 밖으로 빼기.
plt.show()
plt.savefig("data/ch13.png") #그래프를 파일로 저장하려고 할 때.
```
