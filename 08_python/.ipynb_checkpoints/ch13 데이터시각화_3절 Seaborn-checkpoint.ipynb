## 3.2 미적 속성 설정하기

1) 스타일 지정

```python
sns.set(style="darkgrid")
# style : white(기본), darkgrid, whitegrid,dark 등
ax=sns.scatterplot(x="petal_length",y='petal_width',data=iris)
ax
```

2) 컬러팔레트
https://seaborn.pydata.org/generated/seaborn.color_palette.html#seaborn.color_palette
https://matplotlib.org/stable/tutorials/colors/colormaps.html
https://seaborn.pydata.org/tutorial/color_palettes.html
http://hleecaster.com/wp-content/uploads/2019/12/article2_image9.png

```python
sns.set(palette="pastel",style='white')
#brite/deep/muted/pastel 등...
sns.scatterplot(x='petal_length',y='petal_width',data=iris,hue='species')
```

3) 컨텍스트 지정(비율)

```python
sns.set_context("notebook",font_scale=1.5)
#notebook(기본값, 1배율), paper(0.8),talk(1.3), poster(1.6)
ax=sns.scatterplot(x='petal_length',y='petal_width',data=iris)
```

```python
sns.set(palette="muted",style='white')
sns.set_context("notebook",font_scale=1.5, rc={'lines.linewidth':3})
#notebook(기본값, 1배율), paper(0.8),talk(1.3), poster(1.6)
ax=sns.lineplot(x='petal_length',y='petal_width',data=iris)
ax=sns.scatterplot(x='petal_length',y='petal_width',data=iris)
plt.show()
```

## 3.3 seaborn 함수로 그래프 그리기 

Relational plots : 관계형 그래프

```python
sns.scatterplot(x='petal_length',y='petal_width',data=iris)
plt.show()
```

```python
# hue 속성에 그룹변수를 넣어 그룹화하여 그룹별 다른 색상 다른 style
#ris.describe(exclude='float64')
sns.scatterplot(data=iris,x="petal_length",y="petal_width",
               hue='species',palette="bright",style='species') #여기서 적용한 팔레트는 여기서만 적용된다.
sns.scatterplot()
```

```python
sns.set_context("notebook",font_scale=1, rc={'lines.linewidth':3})
sns.lineplot(x='petal_length',y='petal_width',data=iris,hue='species')
sns.scatterplot(x='petal_length',y='petal_width',data=iris,hue='species')
plt.show
```

```python
sns.scatterplot(x='petal_length',y='petal_width',data=iris,hue='species')
sns.lineplot(x='petal_length',y='petal_width',data=iris)
```

```python
fig,axes=plt.subplots(ncols=2,figsize=(8,5))
plt.subplots_adjust(wspace=0.2)
sns.scatterplot(x='petal_length',y='petal_width',data=iris,hue='species',ax=axes[0])
sns.lineplot(x='petal_length',y='petal_width',data=iris,ax=axes[1])
plt.show()
```

```python
sns.set_context("notebook",font_scale=1, rc={'lines.linewidth':3})
sns.lineplot(x='petal_length',y='petal_width',data=iris,hue='species')
sns.scatterplot(x='petal_length',y='petal_width',data=iris,hue='species')
plt.show
```

```python
sns.lineplot(x="petal_length", y="petal_width", data=iris, hue="species")
```

```python
# lineplot을 서브플롯으로 표현: replot

sns.relplot(x="petal_length",y='petal_width',data=iris,hue='species',
           kind='line', #기본값은 scatter
           col='species',
           col_wrap=2) #한 줄에 표현될 서브플롯의 수
```

2) Categorical plots 범주형 그래프

```python
# x축이 범주형 데이터일 경우 scatterplot은 적합하지 않아 다른 것들...
plt.figure(figsize=(12,3))
ax = sns.scatterplot(x="species", y="petal_width", data=iris)
```

```python
# x축이 범주형일 때 쓰는 산점도 1.
plt.figure(figsize=(12,3))
ax=sns.stripplot(x="species",y="petal_length",data=iris)
```

```python
# x축이 범주형일 때 쓰는 산점도 2 : 산점도가 중첩되지 않게 그려지는 장점이 잇다.
plt.figure(figsize=(10,5))
ax=sns.swarmplot(x="species",y="petal_width",data=iris)
```

```python
titanic.head()
```

```python
# 성별 survived의 평균을 막대그래프로
a = sns.barplot(x="sex", y="survived", data=titanic)
```

```python
#class별로 따로 성별 생존률 평균
a=sns.barplot(x="sex",y="survived",data=titanic,hue="class",ci=None)
a
```

```python
#barplot을 서브플롯으로: catplot
```

```python
a=sns.catplot(x="sex",y="survived",data=titanic,hue="class",ci=None,
             kind="bar",col="class")
a

#ci=신뢰구간
```

```python
#deck별 관측수를 막대그래프로.
```

```python
sns.countplot(x='deck',data=titanic,palette="Greens_d")
```

```python
# class별 survived 값을 sex 별로 pointplot
plt.figure(figsize=(10,5))
sns.pointplot(x="class",y="survived",hue="sex",data=titanic,palette={"male":"b","female":"r"},
             markers=["^","o"],
             linestyles=['-','--'])
```

```python
#boxplot,vaiolinplot, boxenplot
sns.boxplot(y="sex",x="age",data=titanic,hue="survived")
```

```python
sns.violinplot(y="sex",x="age",data=titanic,hue="survived")
plt.legend(loc='center right',bbox_to_anchor=(1.25,0.5),ncol=1)
```

```python
sns.boxenplot(y="sex",x="age",data=titanic,hue="survived")
plt.legend(loc='center right',bbox_to_anchor=(1.25,0.5),ncol=1)
```

```python
sns.pairplot(data=iris,hue='species',palette='pastel',markers=['o','s','D'])
```

4) 회귀 그래프

```python
sns.lmplot(x="petal_width",y="petal_length",data=iris)
sns.scatterplot(x="petal_width",y="petal_length",data=iris)
```

```python
sns.lmplot(x="petal_width",y="petal_length",data=iris,hue="species")
sns.scatterplot(x="petal_width",y="petal_length",data=iris)

#hue=group by 하고싶은 그룹
```

```python
#회귀모형을 마드는 함수: lmplot, regplot
#regplot단점: hue를 못씀
sns.regplot(x="petal_width",y="petal_length",data=iris)
```

```python
sns.lmplot(x="petal_width",y="petal_length",data=iris,hue='species',col='species')
```
