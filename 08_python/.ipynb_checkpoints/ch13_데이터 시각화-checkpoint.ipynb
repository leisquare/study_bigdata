2) subplots() 함수로 서브플롯 추가하기 

```python
x=np.arange(0,10,0.01)
x.shape
```

```python
# 방법1(2행 2열 서브플롯)
plt.subplot(2,2,1); plt.plot(x,np.sin(x))
plt.subplot(2,2,2); plt.plot(x,np.cos(x))
plt.subplot(2,2,3); plt.plot(x,np.tanh(x))
plt.subplot(2,2,4); plt.plot(x,np.sin(x)*np.cos(x))
plt.show()
```

```python
# 방법 2 ( 2행 2열 서브플롯)
fit,axes=plt.subplots(nrows=2,ncols=2,figsize=(10,6))
axes[0,0].plot(x,np.sin(x),color='r',linewidth=10)
axes[0,0].plot([0,1,2,3],[-1,-0.5,0,0.5],'k:')
axes[0,1].plot(x,np.cos(x),'c--') # c(청록) --(dashed)
axes[1,0].plot(x,np.tanh(x),'g')
axes[1,1].plot(x,np.sin(x)*np.cos(x),'r-.')
```

```python
# 방법 3 (2행 2열 서브플롯)
x= np.arange(0,7,0.01)
x.shape
```

```python
#def sin_cos(x):
#    return np.sin(x)*np.cos(x)
#func_list=[np.sin,np.cos,np.tanh,sin_cos]

func_list=[np.sin,np.cos,np.tanh,lambda x : np.sin(x)*np.cos(x)]
fmt_list=['k:','c--','g','r-.']
```

```python
fit,axes=plt.subplots(2,2,figsize=(10,6))
#2차원 axes를 list(axes.flat) 이용해서 중첩 for문 안 씀
for i, ax in enumerate(axes.flat):
    ax.plot(x,func_list[i](x),fmt_list[i])
```

```python
fig,axes=plt.subplots(ncols=4,figsize=(12,3))
for i, ax in enumerate(axes): #flat 안 써도 이미 1차원
    ax.plot(x,func_list[i](x),fmt_list[i])
```

```python
fig, axes=plt.subplots(nrows=4,figsize=(10,8))
for i, ax in enumerate(axes):
    ax.plot(x,func_list[i](x),fmt_list[i])
```

```python
fig,axes=plt.subplots(2,2,figsize=(8,5))
fig.suptitle("figure sample plots")
axes[0,0].plot([1,2,3,4],'ro-') # r 빨강, o 동그라미마크, - 실선
axes[0,1].plot(np.random.randn(4,10),
              np.random.randn(4,10),'cs-.') #cyan 청록, s 네모, -. 대시닷
axes[1,0].plot(np.linspace(0,5),np.cos(np.linspace(0,5))) 
axes[1,1].plot([3,6],[3,5],'b^:') #파랑./세모/점선
axes[1,1].plot([4,5],[5,4],'kx--') #검정엑스대시
axes
plt.show()
```

```python
import seaborn as sns
iris_df=sns.load_dataset("iris")
```

```python
iris_df.plot(kind='line',figsize=(12,4))
plt.legend(bbox_to_anchor=(1.01,1),loc=2,borderaxespad=0)
plt.show
```

```python
iris_df.plot(kind='box',figsize=(10,5),title="IRIS box plot",ylabel='y-value',grid=True,
            fontsize=20,rot=30)
#plt.legend(bbox_to_anchor=(1.01,1),loc=2,borderaxespad=0)
plt.show()
```

```python
#iris_df에서 상관관계가 가장 높은 두 변수의 xcatter plot을 그리시오.
iris_df.corr()
```

```python
iris_df.plot(kind="scatter",x="petal_width",y="petal_length",figsize=(5,5))
plt.show()
```

```python
iris_df.groupby(iris_df["species"]).plot(kind="box")
plt.show()
```

```python
iris_df.groupby(iris_df["species"]).plot(kind="box",
                                         subplots=True,
                                         sharey=True,
                                         layout=(3,5))
plt.show()
```

4) pyplot.scatter() : 산점도 

```python
np.random.seed(2021)
N=50
x=np.random.rand(N) # 난수 실수 50개
y=np.random.rand(N)
colors= np.random.rand(N)
area=np.random.rand(N)*30**2
plt.scatter(x,y,
            s=area, #마커 크기
            c=colors, #점 색깔
            alpha=0.5) #%투명도 0 (투명) 1(불투명)
```

## 2.5 다양한 그래프 그리기 예
1) fill(),fill_between(),scatter()<br>
2) bar(),axvline(),axhline()...<br>
3) hist(),boxplot(),violineplot()<br>

```python
x=np.linspace(0,10)
y=np.cos(x)
fig,axes = plt.subplots(2,2)
axes[0,0].scatter(x,y,s=5,c='r',marker='o')
axes[0,1].fill(x,y,c='b')
axes[1,1].fill_between(x,y,color='r')
axes[1,1].plot([-1,5],[-1,1],'ro-')
```

```python
fig,axes=plt.subplots(2,2)
axes[0,0].bar([1,2,3],[3,4,5], color='r')
axes[0,0].plot([1,2,3],[3,4,5], 'k')
axes[0,1].bar([0.5,1,1.5],[0,1,2])
axes[0,1].plot([0.5,1,1.5],[0,1,2], 'g')
axes[1,0].axvline(0.5)
axes[1,1].axhline(0.75)
plt.show
```

```python
fig,axes = plt.subplots(2,2)
#빈도표
axes[0,0].hist(iris_df["sepal_length"],bins=20,color='r')
#누적빈도표=cumulative=True
axes[0,1].hist(iris_df["sepal_length"],bins=20,cumulative=True)
#옆으로 누운!
axes[1,0].hist(iris_df["sepal_length"],bins=10,orientation='horizontal')
axes[1,1].hist(iris_df["sepal_length"],bins=20,histtype='step')
```

```python
fig,axes=plt.subplots(1,2)
axes[0].boxplot(iris_df['sepal_length'])
axes[1].violinplot(iris_df['sepal_length'])
plt.show()
```

## 2.6 그래프 커스터마이징

1) linestyle,linewidth

```python
x=np.linspace(0,10,100)
y=np.cos(x)
```

```python
fig,axes = plt.subplots(2,2)
axes[0,0].plot(x,y,linewidth=2,color='r')
axes[0,1].plot(x,y,ls='dotted',linewidth=5,color='purple')
axes[1,0].plot(x,y,ls='--',linewidth=1)
axes[1,1].plot(x,y,ls='-.',color='b')  # ls=linestyle
```

2) text(), annotation (어노테이션은 화살표 포함)

```python
x=np.linspace(0,10,100)
y=np.cos(x)
```

```python
fig,axes=plt.subplots(1,2,figsize=(8,5))
axes[0].scatter(x,y,marker='.')
axes[0].text(2,0,'sample txt',style='italic')
axes[1].scatter(x,y,marker='*',color='r')
axes[1].annotate('Sine',xytext=(2,0.75),#글자가 있을 위치
                xy=(5,0.5), #화살표가 가리킬 위치
                arrowprops=dict(arrowstyle='->',
                               connectionstyle='angle3'))
plt.show()
```

```python
# 한글 기본폰트 설정
plt.rc("font",family="Malgun Gothic") #윈도우즈에서설정하기
```

```python
fig,axes=plt.subplots(1,2,figsize=(8,5))
axes[0].scatter(x,y,marker='.')
axes[0].text(2,0,'ㅁㄴㅇㄹ',style='italic')
axes[1].scatter(x,y,marker='*',color='r')
axes[1].annotate('Sine',xytext=(2,0.75),#글자가 있을 위치
                xy=(5,0.5), #화살표가 가리킬 위치
                arrowprops=dict(arrowstyle='->',
                               connectionstyle='angle3'))
plt.show()
```
