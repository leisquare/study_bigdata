```python
import re
data= "이름:홍길동, 주소:서울 전화 : 02-716-1006, 특징:번적번쩍"
print(data)
```

```python
phone_pattern=r'[0-9]{2,3}-[0-9]{3,4}-[0-9]{4}' #숫자는 \d로도 표현가능
phone= re.findall(phone_pattern,data)
print(phone)
```

```python
phone_pattern = r'[\d]{2,3}-[\d]{3,4}-[\d]{4}'
phone = re.findall(phone_pattern, data)
print('data에서 추출된 전화번호 ', phone)
```
