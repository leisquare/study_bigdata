```python
#시작 전 설정
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline 
%config InlineBackend.figure_format='retina'
import pandas as pd
import numpy as np
#원하는 한글폰트 지정
plt.rc("font",family="Malgun Gothic")
plt.rc("axes",unicode_minus=False) #Y축
#경고메세지 안보이게
import warnings
warnings.filterwarnings(action='ignore')
import seaborn as sns
custom_palette = sns.color_palette("RdGy", 9)
sns.palplot(custom_palette)
```

```python
import pandas as pd
lvalues=pd.read_csv('E:/kim_jiwon/src/Project_02/data/선호하는_가치_20210412165024.csv',sep=',',encoding='cp949')
lvalues
```

```python
# 일단 시계열/문항/항목/전체만 남긴다.
# 전체전체
lvalues1=lvalues.iloc[0:3,2:]
lvalues1=lvalues1.T.reset_index()
lvalues1.columns=['년도','문항','항목','응답수치']
lvalues1["년도"]= lvalues1["년도"].apply(lambda x:int(x.split(".")[0]))
lvalues1.응답수치=lvalues1.응답수치.astype(float)
lvalues1
```

```python
lvalues1.info()
```

```python
#문항별로 묶고 항목별로 나눠야? 
sns.catplot(data=lvalues1, x="년도",y="응답수치", kind="point", col="문항", hue="항목",col_wrap=1,
                      height=3, aspect=1, linewidth=2,ci=None) 

plt.legend(bbox_to_anchor=(1.01,1.0),loc=0,borderaxespad=0)
#plt.savefig('E:/kim_jiwon/src/project_02/img/주관적_정신건강_수준_문항별항목별.png')
```

```python
plt.title('선호하는 가치_일과 여가')
sns.pointplot(data=lvalues1[lvalues1["문항"]=='일과 여가 중시 선호도'], x="년도",y="응답수치", kind="bar", col="문항", hue="항목",col_wrap=1,
                      height=3, aspect=3, linewidth=2,ci=None) # aspect은 세로:가로가 1:5
plt.legend(bbox_to_anchor=(1.01,1.0),loc=0,borderaxespad=0)

plt.savefig('E:/kim_jiwon/src/project_02/img/선호가치_일과 여가.png')
```
