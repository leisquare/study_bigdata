```python
#시작 전 설정
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline 
%config InlineBackend.figure_format='retina'
import pandas as pd
import numpy as np
#원하는 한글폰트 지정
plt.rc("font",family="Malgun Gothic")
plt.rc("axes",unicode_minus=False) #Y축
#경고메세지 안보이게
import warnings
warnings.filterwarnings(action='ignore')
import seaborn as sns
custom_palette = sns.color_palette("RdGy", 9)
sns.palplot(custom_palette)
```

```python
med_2018=pd.read_excel('E:/kim_jiwon/src/Project_02/data/1_약효분류별다빈도현황(본체)(2017-2018).xls')
med_2021=pd.read_excel('E:/kim_jiwon/src/Project_02/data/1_약효분류별다빈도현황(본체).xls')
med_2021.head(10)
```

```python
med_2021
```

```python
#5행 내용을 컬럼으로 설정
med_2018.columns=med_2018.iloc[5,:]
med_2021.columns=med_2021.iloc[5,:]
#6행, 117.119번약물행 추리기
cd2018_1=med_2018.iloc[6,:].to_frame().T
cd2018_2=med_2018[med_2018.iloc[:,0]=='117']
cd2018_3=med_2018[med_2018.iloc[:,0]=='119']
cd2021_1=med_2021.iloc[6,:].to_frame().T
cd2021_2=med_2021[med_2021.iloc[:,0]=='117']
cd2021_3=med_2021[med_2021.iloc[:,0]=='119']
# '수량' 열 추리기

med2018=pd.concat([cd2018_1,cd2018_2,cd2018_3]).reset_index(drop=True)
med2021=pd.concat([cd2021_1,cd2021_2,cd2021_3]).reset_index(drop=True)
```

```python
med2018
```

```python
med2021
```

```python
med2018.rename(columns= {np.nan:'약효분류코드'}, inplace = True)
med2018=med2018.T

med2021.rename(columns= {np.nan:'약효분류코드'}, inplace = True)
med2021=med2021.T
```

```python
med2018.iloc[1,:]

med2021.iloc[1,:]
med2021.head(5)
```

```python
med2018=med2018.iloc[2:,:].reset_index()
med2021=med2021.iloc[2:,:].reset_index()
```

```python
med2018.columns=['기간','항목','정신신경용제','기타 중추신경용']
med2021.columns=['기간','항목','정신신경용제','기타 중추신경용']

#형변환
med2018.replace(',','', regex=True, inplace=True)
med2021.replace(',','', regex=True, inplace=True)

med2018[['정신신경용제','기타 중추신경용']] = med2018[['정신신경용제','기타 중추신경용']].astype(int)
med2021[['정신신경용제','기타 중추신경용']] = med2021[['정신신경용제','기타 중추신경용']].astype(int)
```

```python
medicine=pd.concat([med2018,med2021],axis=0)
```

```python
medicine
```

```python
med_amount=medicine[medicine['항목']=='수량']
med_price=medicine[medicine['항목']=='사용금액']
```

```python
med_amount
```

```python
fig=plt.figure(figsize=(20,8))
ax = fig.add_axes([0.1,0.1,0.7,0.7])
plt.bar(x="기간", height="정신신경용제", width=0.4, align='edge', data=med_amount)
plt.plot("기간", "정신신경용제",data=med_amount,linewidth=3, color='Blue')
plt.bar(x="기간", height="정신신경용제", width=0.4, align='edge', data=med_price)
plt.plot("기간", "정신신경용제",data=med_price,linewidth=3)
ax.set_title('약물사용량추이확인(정신신경용)', fontsize=20)
ax.set_xlabel('기간')
ax.set_ylabel('수량')
plt.legend(bbox_to_anchor=(1.01,1.0),loc=0,borderaxespad=0)

ax.tick_params(axis='x',labelsize=13,rotation=45)
ax.tick_params(axis='y',labelsize=15)
#plt.ylim([100000,580000])
plt.savefig('E:/kim_jiwon/src/project_02/graph/약물사용량 추이 확인(정신신경용).png')
```

```python
fig=plt.figure(figsize=(20,8))
ax = fig.add_axes([0.1,0.1,0.7,0.7])
plt.bar(x="기간", height="기타 중추신경용", width=0.4, align='edge', data=med_amount)
plt.plot("기간", "기타 중추신경용",data=med_amount,linewidth=3, color='Blue')
plt.bar(x="기간", height="기타 중추신경용", width=0.4, align='edge', data=med_price)
plt.plot("기간", "기타 중추신경용",data=med_price,linewidth=3)
ax.set_title('약물사용량추이확인(중추신경용)', fontsize=20)
ax.set_xlabel('기간')
ax.set_ylabel('수량')
plt.legend(bbox_to_anchor=(1.01,1.0),loc=0,borderaxespad=0)

ax.tick_params(axis='x',labelsize=13,rotation=45)
ax.tick_params(axis='y',labelsize=15)
#plt.ylim([100000,580000])
plt.savefig('E:/kim_jiwon/src/project_02/graph/약물사용량추이확인(중추신경).png')
```
