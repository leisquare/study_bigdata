```python
from selenium import webdriver 
import time


driver = webdriver.Chrome('chromedriver.exe')
driver.get('https://news.naver.com/')
#동작멈춤
time.sleep(0.5) 
```

```python
keyword = '코로나블루' #코로나 라는 문자열을 keyword라는 변수에 담습니다. 
xpath = '//*[@id="lnb.searchForm"]/fieldset/input[1]' 
element = driver.find_element_by_xpath(xpath) #웹요소를 xpath를 이용해 특정하여 줍니다. 
element.send_keys(keyword+ '\n') # 위에서 특정된 웹요소에 '코로나'+엔터를 입력하여 줍니다. 
time.sleep(0.5) # 웹페이지를 이동하고 웹페이지 로딩을 0.5초 기다립니다.

driver.switch_to.window(driver.window_handles[1])
articlebox = driver.find_element_by_class_name('list_news')
articles = articlebox.find_elements_by_class_name('bx')
for article in articles:
    title = article.find_element_by_class_name('news_tit').text
    print(title)
```

```python
from selenium import webdriver 
import time
import csv

driver = webdriver.Chrome('chromedriver.exe')
driver.get('https://news.naver.com/')
time.sleep(0.5)

keyword = '코로나 블루'
xpath = '//*[@id="lnb.searchForm"]/fieldset/input[1]'
element = driver.find_element_by_xpath(xpath)
element.send_keys(keyword+ '\n')
time.sleep(0.5)

driver.switch_to.window(driver.window_handles[1])

results=[]
page = 1
while True:
    pagebox = driver.find_element_by_class_name('sc_page_inner')
    pagebox.find_element_by_link_text(str(page)).click()
    
    driver.switch_to.window(driver.window_handles[1])
    articlebox = driver.find_element_by_class_name('list_news')
    articles = articlebox.find_elements_by_class_name('bx')
    for article in articles:
        press = article.find_element_by_class_name('info.press').text
        title = article.find_element_by_class_name('news_tit').text
        print(title)
        results.append(title)
        
        #if 'https://news.naver.com/' not in driver.current_url:
         #   driver.close()
          #  driver.switch_to.window(driver.window_handles[1])
           # continue

        driver.switch_to.window(driver.window_handles[1])
        
    page += 1
    if page > 399 : break
        
#with open('E:/kim_jiwon/src/Project_02/data/complete.csv','w',newline='', encoding = 'utf-8')as f:
#    f.write(results)
```

```python
def ListToCsv(dataList, csv_filename) :
    with open(csv_filename, 'w', newline='\n', encoding='utf-8') as CSVfile: # '' or "\n"
#         thedatawriter = csv.writer(CSVfile)
        
        for row in dataList:
            CSVfile.write(row+'\n')

# list to CSV
csv_filename = 'comp.txt'
ListToCsv(results, csv_filename)
```

```python
analy=[]
f = open("comp.txt", 'r', encoding='utf-8')
while True:
    line = f.readline()
    analy.append(line)
    if not line: break
    print(line)
    
f.close()
```

```python
# 뉴스 기사 link들 list 가져오기
#https://fs.jtbc.joins.com//RSS/economy.xml
import requests
from bs4 import BeautifulSoup
rss_url = "https://fs.jtbc.joins.com//RSS/economy.xml"
jtbc_economy = requests.get(rss_url)
economy_soup = BeautifulSoup(jtbc_economy.content,"xml")
economy_soup
link_list = economy_soup.select('item > link')
link_list = [link.text for link in link_list]
link_list
```

```python
# 뉴스 기사를 명사들만 뽑아 워드 임베딩하기 좋게 만들기.
from konlpy.tag import Kkma
kkma = Kkma()
cont=[]

for a in analy:
    cont.append(kkma.nouns(a))
```

```python
cont
```

```python
# 워드 임베딩하기
from gensim.models import Word2Vec
model=Word2Vec(cont,size=5000,window=5,min_count=2,workers=-1)
```

```python
model.wv.most_similar("우울")
```

```python
from konlpy.tag import Hannanum
from konlpy.tag import Kkma
from konlpy.tag import Komoran
komoran = Komoran(max_heap_size=2048)

word_list = komoran.nouns("%r" % analy) # s가 아닌 r을 써야한다.
word_list
```

```python
print('명사 출현 수: ',len(word_list))
```

```python
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
```

```python
wordc = WordCloud()
text = ' '.join(word_list)
wordc.generate(text)
plt.figure()
plt.imshow(wordc,interpolation = 'bilinear')
```

```python
spwords = set(STOPWORDS)  # 제외할 단어
spwords.add('골든')  # 제외하고 싶은 단어 추가
spwords.add('블루') 
spwords.add('코로나')
spwords.add('블루 코로나') 
spwords.add('코로나') 
spwords.add('코로나 블루') 

wordc = WordCloud(background_color='white',
                 max_words=300,
                 stopwords=spwords,
                 font_path="c:/Windows/fonts/malgun.ttf",
                 relative_scaling=0.2,width=1600, height=800)
wordc.generate(text)
```
