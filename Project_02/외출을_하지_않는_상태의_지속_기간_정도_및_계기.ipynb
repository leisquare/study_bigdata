```python
#시작 전 설정
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline 
%config InlineBackend.figure_format='retina'
import pandas as pd
import numpy as np
#원하는 한글폰트 지정
plt.rc("font",family="Malgun Gothic")
plt.rc("axes",unicode_minus=False) #Y축
#경고메세지 안보이게
import warnings
warnings.filterwarnings(action='ignore')
import seaborn as sns
custom_palette = sns.color_palette("RdGy", 9)
sns.palplot(custom_palette)
```

```python
import pandas as pd
inside=pd.read_csv('E:/kim_jiwon/src/Project_02/data/외출을_하지_않는_상태의_지속_기간_정도_및_계기_20210412161948.csv',sep=',',encoding='cp949')
inside
```

```python
inside1=inside.iloc[0:3,2:]
inside1=inside1.T.reset_index()
inside1.columns=['년도','문항','항목','응답수치']
inside1["년도"]= inside1["년도"].apply(lambda x:int(x.split(".")[0]))
inside1.응답수치=inside1.응답수치.astype(float)
inside1
```

```python
inside_l=inside1[inside1['문항']=='외출을 하지 않은 기간']
inside_l
inside_r=inside1[inside1['문항']=='외출을 하지 않게 된 계기']
inside_r
```

```python
plt.title('외출하지 않은 기간: 전체')
save2=sns.pointplot(data=inside_l, x="년도",y="응답수치", kind="point", hue="항목",ci=None) # aspect은 세로:가로가 1:5
save2
plt.legend(bbox_to_anchor=(1.01,1.0),loc=0,borderaxespad=0)

plt.savefig('E:/kim_jiwon/src/project_02/img/외출하지 않은 기간.png')
```

```python
plt.title('외출하지 않게 된 계기: 전체')
save2=sns.pointplot(data=inside_r, x="년도",y="응답수치", kind="point", hue="항목",ci=None) # aspect은 세로:가로가 1:5
save2
plt.legend(bbox_to_anchor=(1.01,1.0),loc=0,borderaxespad=0)

plt.savefig('E:/kim_jiwon/src/project_02/img/외출하지 않게 된 계기_전체.png')
```

```python
inside2=inside.iloc[[0,1,3,4],2:]
inside2=inside2.T.reset_index()
inside2.columns=['년도','문항','항목','남','녀']
inside2["년도"]= inside2["년도"].apply(lambda x:int(x.split(".")[0]))
inside2.남=inside2.남.astype(float)
inside2.녀=inside2.녀.astype(float)
inside2
```
