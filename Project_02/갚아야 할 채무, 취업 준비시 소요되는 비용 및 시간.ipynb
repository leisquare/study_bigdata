```python
#시작 전 설정
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline 
%config InlineBackend.figure_format='retina'
import pandas as pd
import numpy as np
#원하는 한글폰트 지정
plt.rc("font",family="Malgun Gothic")
plt.rc("axes",unicode_minus=False) #Y축
#경고메세지 안보이게
import warnings
warnings.filterwarnings(action='ignore')
import seaborn as sns
```

```python
import pandas as pd
debt=pd.read_csv('E:/kim_jiwon/src/Project_02/data/갚아야_할_채무_여부_20210412164556.csv',sep=',',encoding='cp949')
debt
```

```python
debt1=debt.iloc[0:4,1:]
debt1=debt1.T.reset_index()
debt1.columns=['년도','문항','소계','남자','여자']
debt1["년도"]= debt1["년도"].apply(lambda x:str(x.split(".")[0]))
debt1=debt1.loc[1:,]
debt1.loc[:,'소계':'여자']=debt1.loc[:,'소계':'여자'].astype(float)
debt1
```

```python
plt.title('title')
sns.pointplot(data=debt1[debt1['문항']=='있다'], x="년도",y="소계", kind="point", hue="문항",ci=None)
sns.pointplot(data=debt1[debt1['문항']=='있다'], x="년도",y="남자", kind="point", hue="문항",ci=None)
sns.pointplot(data=debt1[debt1['문항']=='있다'], x="년도",y="여자", kind="point", hue="문항",ci=None)
plt.legend(bbox_to_anchor=(1.01,1.0),loc=0,borderaxespad=0)
#plt.savefig('E:/kim_jiwon/src/project_02/img/외출하지 않은 기간_여성.png')
```

```python
import pandas as pd
for_job=pd.read_csv('E:/kim_jiwon/src/Project_02/data/취업_준비_시_소요되는_비용_및_시간_20210412164155.csv',sep=',',encoding='cp949')
for_job
```

```python
for_job1=for_job.iloc[0:4,1:]
for_job1=for_job1.T.reset_index()
for_job1.columns=['년도','문항','소계','남자','여자']
for_job1["년도"]= for_job1["년도"].apply(lambda x:str(x.split(".")[0]))
for_job1=for_job1.loc[1:,]
for_job1.loc[:,'소계':'여자']=for_job1.loc[:,'소계':'여자'].astype(float)
for_job1
```
