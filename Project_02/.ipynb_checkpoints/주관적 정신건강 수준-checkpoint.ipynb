```python
#시작 전 설정
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline 
%config InlineBackend.figure_format='retina'
import pandas as pd
import numpy as np
#원하는 한글폰트 지정
plt.rc("font",family="Malgun Gothic")
plt.rc("axes",unicode_minus=False) #Y축
#경고메세지 안보이게
import warnings
warnings.filterwarnings(action='ignore')
import seaborn as sns
```

```python
import pandas as pd
mental=pd.read_csv('E:/kim_jiwon/src/Project_02/data/주관적인_정신건강_수준_20210412155804.csv',sep=',',encoding='cp949')
mental
```

```python
# 일단 시계열/문항/항목/전체만 남긴다.
#4개 행만 남겨보기
mental1=mental.iloc[0:3,2:]
mental1=mental1.T.reset_index()
mental1.columns=['년도','문항','항목','응답수치']
mental1["년도"]= mental1["년도"].apply(lambda x:int(x.split(".")[0]))
mental1
```

```python
mental1.응답수치=mental1.응답수치.astype(float)
```

```python
#가중치 테이블 만들기
mental1["가중치"]= ['0','1','2','3','4','5']*5*4
mental1.가중치=mental1.가중치.astype(float)
mental1["결과값"]=mental1['응답수치']*mental1['가중치']
mental=mental1
```

```python
# 년도별로 나눠야함
#data[data['first_name'] == 'Antonio']

mental2017=mental[mental['년도']==2017]
mental2018=mental[mental['년도']==2018]
mental2019=mental[mental['년도']==2019]
mental2020=mental[mental['년도']==2020]

mental2017
```
