```python
#시작 전 설정
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline 
%config InlineBackend.figure_format='retina'
import pandas as pd
import numpy as np
#원하는 한글폰트 지정
plt.rc("font",family="Malgun Gothic")
plt.rc("axes",unicode_minus=False) #Y축
#경고메세지 안보이게
import warnings
warnings.filterwarnings(action='ignore')
import seaborn as sns
```

```python
import pandas as pd
s_health=pd.read_csv('E:/kim_jiwon/src/Project_02/data/주관적인_신체_건강에_대한_인식_20210412161750.csv',sep=',',encoding='cp949')
s_health
```

```python
s_health1=s_health.iloc[0:2,2:]
s_health1=s_health1.T.reset_index()
s_health1.columns=['년도','항목','응답수치']
s_health1["년도"]= s_health1["년도"].apply(lambda x:int(x.split(".")[0]))
s_health1.응답수치=s_health1.응답수치.astype(float)
s_health1
```

```python
plt.title('주관적 신체 건강에 대한 인식')
save1=sns.pointplot(data=s_health1, x="년도",y="응답수치", kind="point", hue="항목",ci=None) # aspect은 세로:가로가 1:5
save1
plt.legend(bbox_to_anchor=(1.01,1.0),loc=0,borderaxespad=0)

plt.savefig('E:/kim_jiwon/src/project_02/img/주관적 신체건강에 대한 인식.png')
```

```python
# 돌아와서 가중치합산 해볼것
```

```python
#가중치 테이블 만들기
s_health1

s_health1["가중치"]= ['0','1','2','3','4']*3
s_health1.가중치=s_health1.가중치.astype(float)
s_health1["결과값"]=s_health1['응답수치']*s_health1['가중치']
s_health=s_health1
s_health
```

```python
plt.figure(figsize=(8,10))
plt.title('주관적 신체 건강에 대한 인식2')
save2=sns.pointplot(data=s_health, x="항목",y="결과값", kind="bar", hue="년도", aspect=3, height=5) # aspect은 세로:가로가 1:5
save2
plt.legend(bbox_to_anchor=(1.01,1.0),loc=0,borderaxespad=0)

plt.savefig('E:/kim_jiwon/src/project_02/img/주관적 신체건강에 대한 인식2.png')
```

```python
save3=sns.catplot(data=s_health, x="항목",y="결과값", kind='bar',hue="년도", aspect=1, height=5) # aspect은 세로:가로가 1:5
save3
plt.savefig('E:/kim_jiwon/src/project_02/img/주관적 신체건강에 대한 인식3.png')
```
