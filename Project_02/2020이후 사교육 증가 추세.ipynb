```python
#불필요한 행 드롭(code가 들어간 행)
carddata=carddata.loc[:,~carddata.columns.str.contains('code', case=False)] 
carddata
```

```python
# 행 이름 변경
carddata.columns=['접수일자','행정동명','업종명','건수','금액']
pri_edu_pre=carddata
pri_edu_pre
```

```python
test1=pri_edu_pre[pri_edu_pre['업종명']=='보습학원']
test1
```

```python
pri_edu_pre=pri_edu_pre.set_index('업종명')
test2 = pri_edu_pre.loc[['보습학원','예체능학원','학원(회원제형태)','외국어학원','컴퓨터학원','기능학원','학습지교육']]
test2
```

```python
##접수연월 분리
test2['접수일자'] = test2['접수일자'].apply(str)
test2["접수연월"]= test2["접수일자"].apply(lambda x:x[0:6])
test2
```

```python
test3=test2.reset_index()
test3
```

```python
test3.groupby(['업종명','접수연월']).sum()
```

```python
test3=test3.groupby(['업종명','접수연월']).sum()
test3
```

```python
test3=test3.reset_index()
```

```python
test3[test3['업종명']=='학습지교육']
```

```python
#test3.pivot_table(index='접수연월',columns='업종명',values='금액').plot(figsize=(10,8))
fig=plt.figure(figsize=(20,8))
ax = fig.add_axes([0.1,0.1,0.7,0.7])
plt.plot("접수연월","금액", data=test3[test3['업종명']=='학습지교육'], linewidth=3)
plt.plot("접수연월","금액", data=test3[test3['업종명']=='컴퓨터학원'], linewidth=3)
plt.plot("접수연월","금액", data=test3[test3['업종명']=='외국어학원'], linewidth=3)
plt.plot("접수연월","금액", data=test3[test3['업종명']=='예체능학원'], linewidth=3)
plt.plot("접수연월","금액", data=test3[test3['업종명']=='보습학원'], linewidth=3)
plt.plot("접수연월","금액", data=test3[test3['업종명']=='기능학원'], linewidth=3)
plt.plot("접수연월","금액", data=test3[test3['업종명']=='학원(회원제형태)'], linewidth=3)

plt.title('사교육비')

plt.legend(bbox_to_anchor=(1.01,1.0),loc=0,borderaxespad=0)

#plt.bar(x="기간", height="정신신경용제", width=0.4, align='edge', data=med_amount)
#plt.plot("기간", "정신신경용제",data=med_amount,linewidth=3, color='Blue')
#plt.bar(x="기간", height="정신신경용제", width=0.4, align='edge', data=med_price)
#plt.plot("기간", "정신신경용제",data=med_price,linewidth=3)
#ax.set_title('약물사용량추이확인(정신신경용)', fontsize=20)
```
