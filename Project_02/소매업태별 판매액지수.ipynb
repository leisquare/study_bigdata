```python
data3_t=data3[data3['판매매체별']=='계']
data3_t['년도']=data["시점"].apply(lambda x:str(x.split(".")[0]))
data3_t_t=data3_t.groupby('년도').sum()
data3_t_t
```

```python
data3_f=data3_t_t.reset_index().melt(id_vars=['년도'],var_name="품목",value_name="금액")
data3_f
```

```python
fig=plt.figure(figsize=(15,10))
ax = fig.add_axes([0.1,0.1,0.7,0.7])

sns.pointplot(data=data3_f, x="년도",y="금액", kind="point", hue="품목",ci=None)




plt.title('구입금액')

plt.legend(bbox_to_anchor=(1.01,1.0),loc=0,borderaxespad=0)

#plt.savefig('E:/kim_jiwon/src/project_02/img/규칙적인 운동 빈도.png')
```
