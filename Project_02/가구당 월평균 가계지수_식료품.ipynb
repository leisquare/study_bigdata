```python
import pandas as pd
data_food1=pd.read_csv('E:/kim_jiwon/src/Project_02/data/월평균가계지출세부/가구당_월평균_가계지출__전국_1인이상__20210415175626_식료품.csv',sep=',',encoding='cp949')
data_food2=pd.read_csv('E:/kim_jiwon/src/Project_02/data/월평균가계지출세부/가구당_월평균_가계지출__전국_1인이상__20210415175626_식료품2.csv',sep=',',encoding='cp949')
```

```python

#data.rename(index = {'가계지출항목별(1)':'년도'}, inplace = True)
#data_food1 정제
data_food1["가계지출항목별(2)"]=data_food1["가계지출항목별(2)"].apply(lambda x:str(x.split(" ")[0]))
data_food1
```

```python

#data.rename(index = {'가계지출항목별(1)':'년도'}, inplace = True)
#data_food1 정제
data_food1=data_food1.set_index(data_food1.loc[:,"가계지출항목별(2)"])
data_food1=data_food1.loc[:,"2017":"2018"]
data_food1=data_food1.loc["소계":,"2017":"2018"]
data_food1
```

```python

#data_food2 정제
data_food2=data_food2.set_index(data_food2.loc[:,"가계수지항목별"],drop=True)
data_food2=data_food2.loc[:,"2019":"2020"]
data_food2.rename(index = {"01.식료품 · 비주류음료":"소계"}, inplace = True)
data_food2=data_food2.loc["소계":,"2019":"2020"]
data_food2


#data_lodge=pd.read_csv('E:/kim_jiwon/src/Project_02/data/월평균가계지출세부/가구당_월평균_가계지출__전국_1인이상__20210415175626_음식숙박.csv',sep=',',encoding='cp949')


```

```python
test1=pd.concat([data_food1,data_food2],axis=1)
test1
```

```python
#소계를 제거한 항목 만들기
test1_cases=test1.iloc[1:,:]
```

```python
test1_cases=test1_cases.reset_index()
test1_cases.rename(columns={'index':"항목"},inplace=True)
test1_cases
```

```python
test_cases=test1_cases.melt(id_vars=["항목"],var_name="년도",value_name="가계지수")
test_cases
```

```python
test_cases[['가계지수']] = test_cases[['가계지수']].astype(int)
test_cases.info(0)
```

```python
test_cases.pivot_table(index='년도',columns='항목',values='가계지수').plot(figsize=(10,8))
plt.legend(bbox_to_anchor=(1.01,1),loc=0,borderaxespad=0)
plt.savefig('E:/kim_jiwon/src/project_02/graph/가구당 월평균 가계지수 변화_식료품.png')
```
