```python
#시작 전 설정
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline 
%config InlineBackend.figure_format='retina'
import pandas as pd
import numpy as np
#원하는 한글폰트 지정
plt.rc("font",family="Malgun Gothic")
plt.rc("axes",unicode_minus=False) #Y축
#경고메세지 안보이게
import warnings
warnings.filterwarnings(action='ignore')
import seaborn as sns
custom_palette = sns.color_palette("RdGy", 9)
sns.palplot(custom_palette)
```

```python
import pandas as pd
rsport=pd.read_csv('E:/kim_jiwon/src/Project_02/data/규칙적인_운동_여부_및_빈도_20210412161706.csv',sep=',',encoding='cp949')
rsport
```

```python
# 일단 시계열/문항/항목/전체만 남긴다.
#4개 행만 남겨보기
rsport1=rsport.iloc[0:3,2:]
rsport1=rsport1.T.reset_index()
rsport1.columns=['년도','문항','항목','응답수치']
rsport1["년도"]= rsport1["년도"].apply(lambda x:int(x.split(".")[0]))
rsport1.응답수치=rsport1.응답수치.astype(float)
rsport1
```

```python
rsport_yn=rsport1[rsport1['문항']=='규칙적인 운동 여부']
rsport_yn
```

```python
rsport_f=rsport1[rsport1['문항']=='규칙적인 운동 빈도']
rsport_f
```

```python
#rsportf에 가중치 테이블 만들기
rsport_f["가중치"]= ['0','1','2','3','4','5','6']*4
rsport_f.가중치=rsport_f.가중치.astype(float)
rsport_f["결과값"]=rsport_f['응답수치']*rsport_f['가중치']
rsport_f
```

```python
plt.title('규칙적인 운동 빈도')
save1=sns.pointplot(data=rsport_f, x="년도",y="응답수치", kind="point", hue="항목",ci=None) # aspect은 세로:가로가 1:5
save1
plt.legend(bbox_to_anchor=(1.01,1.0),loc=0,borderaxespad=0)

plt.savefig('E:/kim_jiwon/src/project_02/img/규칙적인 운동 빈도.png')
```

```python
rsport_yn_y=rsport_yn[rsport_yn['항목']=='예']
rsport_yn_y
```

```python
plt.title('규칙적인 운동 여부')
save2=sns.pointplot(data=rsport_yn_y, x="년도",y="응답수치", kind="point", hue="문항",ci=None) # aspect은 세로:가로가 1:5
save2
plt.legend(bbox_to_anchor=(1.01,1.0),loc=0,borderaxespad=0)

plt.savefig('E:/kim_jiwon/src/project_02/img/규칙적인 운동 여부.png')
```
